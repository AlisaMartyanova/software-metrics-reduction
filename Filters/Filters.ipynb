## Reading data

```python
import pandas as pd
from sklearn.preprocessing import normalize
data = pd.read_csv('data.csv')
data = data.drop('Link', axis=1)
data = data.drop('Language', axis=1)
data = data.drop('Owner type', axis=1)
data = data.drop('# Devs closed issues', axis=1)
data = data.drop('# Devs closed pulls', axis=1)
data = data.drop('% merged pulls among closed', axis=1)
data = data[:100]
for i in range(0, len(data)):
    if (data["Average # issues per milestones"][i][2]==' '):
        newData = data["Average # issues per milestones"][i][0]+data["Average # issues per milestones"][i][1]+data["Average # issues per milestones"][i][3]+data["Average # issues per milestones"][i][4]+data["Average # issues per milestones"][i][5];
        data["Average # issues per milestones"][i] = float(newData)
    if (data["Average time to close pull request (days)"][i][2]==' '):
        newData = data["Average time to close pull request (days)"][i][0]+data["Average time to close pull request (days)"][i][1]+data["Average time to close pull request (days)"][i][3]+data["Average time to close pull request (days)"][i][4]+data["Average time to close pull request (days)"][i][5];
        data["Average time to close pull request (days)"][i] = float(newData)
data.tail()
```

### Low variance filter

```python
normalize = normalize(data)
data_scaled = pd.DataFrame(normalize)
data_scaled.var()
```

```python
variance = data_scaled.var()
columns = data.columns
variable = []
THRESHOLD = 0.03

for i in range(0,len(variance)):
    if variance[i]>=THRESHOLD:
        variable.append(columns[i])
print(len(variable))
variable
```

## High correlation filter

```python
correlation = data_scaled.corr(method='spearman')
correlation
```

```python
columns = data.columns
variable = []
THRESHOLD = 0.2
for i in range(0,len(variance)):
    for j in range(0,len(variance)):
        if columns[i]!= columns[j] and (columns[i] not in variable) and (correlation[i][j]>THRESHOLD or correlation[i][j] is not None):
            variable.append(columns[i])
print(variable)
# variable contains columns which are useless
```

```python
new_columns = []
for c in columns:
    if c not in variable:
        new_columns.append(c)
print(len(new_columns))
print(new_columns)
```
